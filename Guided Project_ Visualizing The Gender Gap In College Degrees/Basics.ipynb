```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
#the color of the horizontal line :
c_horiz_line = (127/255, 127/255, 127/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16, 20))


# generate the first column 

#we have seventeen degrees so we changed the rang to 17
for sp in range(0,18,3):
    cat_inedx  = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1) # 6 rows and 3 columns
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_inedx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_inedx]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #enable the y-axis labels at 0 and 100
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[cat_inedx])
    #disable the x-axis labels for all of the line charts with setting the labelbottom 'off'
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom = "off")
    
    #Generate a horizontal line
    ax.axhline(50, c = c_horiz_line, alpha = 0.3 )
    
    if cat_inedx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif cat_inedx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        #turn on the label bottom for the last chart in the first column
        ax.tick_params(labelbottom = "on")
        
#generate the second column
for sp in range(1,16,3):
    cat_inedx  = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1) # 6 rows and 3 columns
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_inedx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_inedx]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #enable the y-axis labels at 0 and 100
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[cat_inedx])
    #disable the x-axis labels for all of the line charts with setting the labelbottom 'off'
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom = "off")
    #Generate a horizontal line
    ax.axhline(50, c= c_horiz_line, alpha = 0.3 )
    
    if cat_inedx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
         #turn on the label bottom for the last chart in the second column
    elif cat_inedx == 4:
        ax.tick_params(labelbottom = "on")
        
        
#generate the third column:

for sp in range(2,20,3):
    cat_inedx  = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1) # 6 rows and 3 columns
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_inedx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_inedx]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #enable the y-axis labels at 0 and 100
    ax.set_yticks([0,100])
    ax.set_title(other_cats[cat_inedx])
    #disable the x-axis labels for all of the line charts with setting the labelbottom 'off'
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")
    #Generate a horizontal line
    ax.axhline(50, c= c_horiz_line, alpha = 0.3 )
    
    if cat_inedx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif cat_inedx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
         #turn on the label bottom for the last chart in the third column
        ax.tick_params(labelbottom = "on")
        
plt.savefigave('gender_degrees.png')
plt.show()
```
