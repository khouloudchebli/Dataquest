```python
#grouped barplot major / sexe
import numpy as np 
majors=recent_grads['Major_category'].unique()
men=[]
women=[]
for m in majors:
    men.append(recent_grads[recent_grads['Major_category']==m]['Men'].sum())
    women.append(recent_grads[recent_grads['Major_category']==m]['Women'].sum())
#fig, ax = plt.subplots()
```

```python
men
```

we select the unique majors, then for each major, we calculate the sum of men, sum of women, then we ploted it side by side

```python
fig, ax = plt.subplots()
ind = np.arange(len(majors))    # the x locations for the groups
plt.xticks(rotation=90 )

width = 0.35         # the width of the bars
p1 = ax.bar(ind, men, width, color='b', bottom=0 )


womenMeans = (145 , 149 , 172 , 165 , 200 )
womenStd = (30 , 25 , 20 , 31 , 22 )
p2 = ax.bar(ind + width, women, width,
           color='g', bottom=0 )
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(majors)

ax.legend((p1[0], p2[0]), ('Men', 'Women'))
ax.autoscale_view()
```
